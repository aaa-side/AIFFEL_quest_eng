QC1-2
회문 분류기

```python
# 회문 분석 함수
def palinddrome_reader(x):
    if x == x[::-1]:
        return True
    else:
        return False


def palinddrome_classifier(word_list):
    palinddrome_dict = {"회문O":[], "회문X":[]}

    for word in word_list:
        if palinddrome_reader(word): # == True는 생략해도 됨
            palinddrome_dict["회문O"].append(word)
        else:
            palinddrome_dict["회문X"].append(word)

            
    pal_count = len(palinddrome_dict["회문O"])
    
    if len(word_list) > 0:
        pal_per = round(pal_count / len(word_list),2)
    else:
        pal_per = 0.0
        
    return palinddrome_dict, pal_count, pal_per


word_list = list(input("앞으로 읽어도 뒤로 읽어도 같은 글자를 입력하세요(공백으로 구분): ").split())
print(palinddrome_classifier(word_list))
```

score_analysis()

```python
scores = [[None,"국어","영어","수학"],["유원영",94,91,89],["설지민",88,96,90],["신윤아",86,88,77],["장윤아",98,76,92]]

def score_analysis(name):
    subjects = scores[0][1:]  # ['국어', '영어', '수학']
    
    class_scores = {sub: [] for sub in subjects}
    student_row = None
    
    for row in scores[1:]:
        if row[0] == name:
            student_row = row
        for idx, sub in enumerate(subjects):
            class_scores[sub].append(row[idx + 1])
            
    # 예외 처리: 리스트에 없는 학생 이름이 들어온 경우
    if not student_row:
        print(f"학생 '{name}'을 찾을 수 없습니다. 이름을 확인해 주세요.")
        return

    class_averages = {}
    for sub in subjects:
        class_averages[sub] = sum(class_scores[sub]) / len(class_scores[sub])
        
    print(f"[{name} 성적 분석]")
    
    student_scores = student_row[1:]
    for idx, sub in enumerate(subjects):
        score = student_scores[idx]
        c_avg = class_averages[sub]
        diff = score - c_avg
        
        print(f"{sub}: {score}점, 학급 평균({c_avg:.2f}점) 대비 {diff:+.2f}")
        
    student_total_avg = sum(student_scores) / len(student_scores)
    class_total_avg = sum(class_averages.values()) / len(class_averages)
    total_diff = student_total_avg - class_total_avg
    
    print()  # 한 줄 띄우기
    print(f"전과목 평균: {student_total_avg:.2f}점, 학급 평균({class_total_avg:.2f}점) 대비 {total_diff:+.2f}")


student_name = input("성적을 분석할 학생의 이름을 입력하세요: ")
score_analysis(student_name)
```
